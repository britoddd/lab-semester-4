```python
from Bio.Seq import Seq
```

# **Protein Synthesis**

```python
dnaSeq = Seq('ATGCCTTAA')
dnaSeq
```

```python
cDNASeq = dnaSeq.complement()
cDNASeq
```

```python
iDNASeq = dnaSeq.reverse_complement()  # dibalik depan belakangnya
iDNASeq
```

```python
# Converting DNA to RNA (manually)
def manual_transcribe(seq):
    mRNA = seq.replace('T', 'U')
    return mRNA

mRNAseq = manual_transcribe(dnaSeq)
mRNAseq
```

```python
dnaSeq.transcribe()
```

```python
mRNAseq.back_transcribe()
```

```python
# Translasi: ngubah dari RNA jadi Protein
amino_acid = mRNAseq.translate()
amino_acid
```

```python
from Bio.SeqUtils import seq1, seq3
```

```python
polyPep = seq3(amino_acid)
polyPep
```

```python
seq1(polyPep)
```
