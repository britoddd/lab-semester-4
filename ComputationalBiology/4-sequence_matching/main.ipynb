# Session 4 - Sequence Alignment
untuk ngebandingin dua sequence berbeda, biar bisa dicari bagian yang mirip utk dicari hubungannya  
*contoh hubungan: fungsi, struktur, evolusi dari ancestor yang sama*

Match -> sama di suatu posisi  
Mismatch -> beda di suatu posisi  
Gap -> space kosong untuk diisikan ke sequence yang lebih pendek untuk improve alignment

Local alignment -> cuma di region tertentu  
Global alignment -> di keseluruhan sequence

```python
from Bio.Seq import Seq
import random
```

```python
ncs = 'GCAT'

seqA = ''.join([random.choice(ncs) for _ in range(20)])
seqB = ''.join([a if random.random() < 0.6 else random.choice(ncs) for a in seqA])

print(seqA)
print(seqB)
```

```python
from Bio import pairwise2 as pw2
from Bio.pairwise2 import format_alignment as fa

globalxx = pw2.align.globalxx(seqA, seqB)
# globalxx = pw2.align.globalxx(seqA, seqB, one_alignment_only=True, score_only=True)

# for a in globalxx:
    # print(fa(a.seqA, a.seqB))
    # print(fa(*a))  # kalo match bakal nambah poin match
    
print(globalxx[0].score)  # biar ngga usah pake one_alignmen_only sama score_only
print(fa(*globalxx[0]))
```

```python
localxx = pw2.align.localxx(seqA, seqB)
print(fa(*localxx[0]))
```

```python
# Match, Mismatch, Gap open, Gap extension
localms = pw2.align.localms(seqA, seqB, 2, -1, -0.5, -0.1)
print(fa(*localms[0]))
```

```python
# Match, Mismatch, Gap open, Gap extension
globalms = pw2.align.globalms(seqA, seqB, 2, -1, -0.5, -0.1)

# for a in globalms:
#     print(fa(*a))
    
print(fa(*globalms[0]))
```

```python
globalxs = pw2.align.globalxs(seqA, seqB, -0.5, -0.1)
print(fa(*globalxs[0]))
```

```python
hammingDist = 0
for a, b in zip(seqA, seqB):
    if a != b:
        hammingDist += 1
        
hammingDist
```

```python
print("  " + seqB)
print("  " + "-" * len(seqB))
for a in seqA:
    line = f"{a}|"
    for b in seqB:
        if a == b:
            line += '■'
        else:
            line += ' '
    
    print(line)
```
