```python
import matplotlib.pyplot as plt
from collections import Counter
```

```python
counterA = Counter(seqA)
counterB = Counter(seqB)
counterC = Counter(seqC)

plt.bar(counterA.keys(), counterA.values())
plt.title("Nucleeotides base (A, C, G, T) frequency for Sequence A")
plt.show()

plt.bar(counterB.keys(), counterB.values())
plt.title("Nucleeotides base (A, C, G, T) frequency for Sequence B")
plt.show()

plt.bar(counterC.keys(), counterC.values())
plt.title("Nucleeotides base (A, C, G, T) frequency for Sequence C")
plt.show()
```
