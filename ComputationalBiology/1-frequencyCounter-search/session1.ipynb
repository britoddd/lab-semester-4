### **Nucleotide**
Intinya nanti matkul ini bakal bermain-main di sini lah ya: G, T, A, dan C

```python
from Bio.Seq import Seq
from Bio.SeqUtils import nt_search
from collections import Counter
from matplotlib import pyplot as plt

seqA = Seq('ATGCGTACGTCAGTCGATCGATCGTAGCTA')
seqB = Seq('GTACGTCCAGTCGTAGCATCGATATGCGTA')
```

```python
# Index
print('Index\t\t:', seqA.index('GC'))  # untuk nyari index dari sequence yang dicari, kalo ga ketemu bakal Error

# Slicing
sliced = seqA[0:5:]  # array[start:end:step(default:1)]
print('Slicing\t\t:', sliced)

# Merging and Joining
print('Merging\t\t:', seqA + seqB)

# Finding Codon -> return index dari yang dicari, kalo ga ketemu bakal return -1
print('Find\t\t:', seqA.find('GTA'))
```

```python
nt_search(str(seqA), 'GTA')  # lebih powerful dibandingkan find bawaan python
```

```python
# Nucleotides Frequency
G_Count = seqA.count("G")
A_Count = seqA.count("A")
C_Count = seqA.count("C")
T_Count = seqA.count("T")

print(f"Nucleotides Frequency\nG: {G_Count}\nA: {A_Count}\nC: {C_Count}\nT: {T_Count}")
```

```python
# Frequency, but using Library
frequency = Counter(seqA)
frequency
```

```python
# Print Output
plt.title("Sequence Base Frequency")
plt.bar(frequency.keys(), frequency.values())
plt.show()
```
