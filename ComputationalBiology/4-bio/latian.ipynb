```python
from Bio.Seq import Seq
from Bio import pairwise2 as pw2
from Bio.pairwise2 import format_alignment as fa

from Levenshtein import distance
```

```python
seqA = Seq('AAGCTTAGCTAGAGCAGTCAGCTAGCTAGCT')
seqB = Seq('TCTGAAGCTAGCTAGTGCATAGCTAGCTAGC')

localxx = pw2.align.localxx(seqA, seqB)
print(fa(*localxx[0]))
```

```python
# Hamming Distance
hamming = 0
for a, b in zip(seqA, seqB):
    if a != b:
        hamming += 1
        
print('Hamming Distance:', hamming)
```

```python
# Levenshtein
levenshtein = distance(seqA, seqB)

print('Levenshtein:', levenshtein)
```

```python
# Map
print('   ' + seqB)
for a in seqA:
    line = f"{a} | "
    for b in seqB:
        if a == b:
            line += '‚Ė†'
        else:
            line += ' '
    print(line)
    
    
```
