```python
from Bio.Seq import Seq
from Bio.SeqUtils import MeltingTemp as mt
from Bio.SeqUtils import gc_fraction
from Bio.SeqUtils import molecular_weight as mw

import numpy as np
import matplotlib.pyplot as plt

seqA = Seq('AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC')
seqB = Seq('CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG')

print(f'Seq A: {seqA}\nSeq B: {seqB}')
```

```python
# Determine the molecular weight of both sequence
mw_seqA = mw(seqA)
mw_seqB = mw(seqB)

print(f'Molecular Weight Seq A: {mw_seqA:.2f}\nMolecular Weight Seq B: {mw_seqB:.2f}')
```

```python
# Refer to the Wallace rule to calculate the melting point of sequence A and sequence B and plot them using Matplotlib library
mt_wallace_A = mt.Tm_Wallace(seqA)
mt_wallace_B = mt.Tm_Wallace(seqB)

temperature = np.linspace(20, 90, 100)

plt.title('Wallace Rule Sequence A & Sequence B')
plt.bar(['Sequence A', 'Sequence B'], [mt_wallace_A, mt_wallace_B], color=['Orange', "Brown"])
plt.show()
```

```python
# Calculate the GC content of sequence A and sequence B
gc_content_A = gc_fraction(seqA) * 100
at_content_A = 100 - gc_content_A

gc_content_B = gc_fraction(seqB) * 100
at_content_B = 100 - gc_content_B

print(f'Total GC Content: {gc_content_A:.2f}%, Total AT Content: {at_content_A:.2f}%')
print(f'Total GC Content: {gc_content_B:.2f}%, Total AT Content: {at_content_B:.2f}%')
```

```python
# Plot the GC and AT content of sequence A and sequence B using Matplotlib library
plt.title('GC & AT Content')

plt.bar([0, 0.7], [gc_content_A, at_content_A], width=.15, color=['Orange', 'Brown'])
plt.bar([0.2, 0.9], [gc_content_B, at_content_B], width=.15, color=['Black', 'Red'])
plt.xticks([0.1, 1.1], ['GC Contents', 'AT Contents'])
plt.ylabel('Percentage %')
plt.show()
```
