```python
from Bio.Seq import Seq
```

### GC & AT Content

```python
seq = Seq('GCTAGCATTAGCTAGCCCGGTTAATTA')
seq
```

```python
def GC_counter(seq):
    count_g = seq.count("G")
    count_c = seq.count("C")
    return (count_g + count_c) / len(seq)

print(f'{GC_counter(seq) * 100}%')
```

```python
from Bio.SeqUtils import gc_fraction
```

```python
gc_contents = gc_fraction(seq) * 100
print(f'GC Contents Sequence: {gc_contents:.2f}%')

# at_contents berarti sisa dari gc_contents yang belom dipake
at_contents = 100 - gc_contents
print(f'AT Contetns Sequence: {at_contents:.2f}%')
```

### Melting Temp (Tm)

```python
from Bio.SeqUtils import MeltingTemp as mt
```

```python
def wallace(seq):
    countC = seq.count('C')
    countG = seq.count('G')
    countA = seq.count('A')
    countT = seq.count('T')
    
    return 2 * (countA + countT) + 4 * (countC + countG)

print(f'Metode Wallace Manual: {wallace(seq)}')
```

```python
# Melting Temp Wallace, GC, Nearest Neighbor

mt_wallace = mt.Tm_Wallace(seq)  # idealnya untuk basa kurang dari 20 aja
print(f'Melting Temp Wallace: {mt_wallace}')

mt_gc = mt.Tm_GC(seq)
print(f'Melting Temp GC: {mt_gc:.2f}')

mt_nn = mt.Tm_NN(seq)
print(f'Melting Temp Nearest Neighbor: {mt_nn:.2f}')
```
