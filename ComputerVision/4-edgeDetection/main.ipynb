# Sesi 4 - Edge Detection
-> cari area dengan perubahan intensitas cahaya yang signifikan

Ada 3 cara untuk melakukan:
- Laplacian
- Sobel X & Y
- Canny

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread('image.jpg')
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.show()
```

```python
grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# untuk image detection pake ini, karena yang diliat cuma intensity
```

Laplacian -> akan cek ke segala arah

Kernel (3x3):\
0  1  0\
1 -4  1\
0  1  0

```python
laplacian_img = cv2.Laplacian(grayscale, cv2.CV_64F)
# 64F: 64 bit floating point, bisa ngesimpen semua penjumlahan positif negatif
#      kalo cuma 8 bit bakal jadi overflow dan negatifnya ngga kesimpen

plt.imshow(laplacian_img, "gray")
plt.show()
```

Sobel

- Sobel x (vertically)\
   1  0  1\
  -2  0  2\
   1  0  1
- Sobel y (horizontally)\
   1 -2  1\
   0  0  0\
   1  2  1

```python
sobel_x = cv2.Sobel(grayscale, cv2.CV_64F, 1, 0)
sobel_y = cv2.Sobel(grayscale, cv2.CV_64F, 0, 1)

# plt.imshow(sobel_x, "gray")
# plt.show()
# plt.imshow(sobel_y, "gray")
# plt.show()

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(sobel_x, "gray")
plt.title("Sobel X")
plt.subplot(1, 2, 2)
plt.imshow(sobel_y, "gray")
plt.title("Sobel Y")
```

# Canny -> akan ada blurnya biar ngga nambah detection yang ngga mutu

```python
canny_img = cv2.Canny(grayscale, 100, 200)  # double threshold
# < 100 -> dibuang
# > 200 -> diambil
# 100-200 -> akan diambil jika terkoneksi dengan pixel yang diambil
plt.imshow(canny_img, 'gray')
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(2, 2, 1)
plt.imshow(np.absolute(laplacian_img), "gray")
plt.subplot(2, 2, 2)
plt.imshow(np.absolute(sobel_x), "gray")
plt.subplot(2, 2, 3)
plt.imshow(np.absolute(sobel_y), "gray")
plt.subplot(2, 2, 4)
plt.imshow(np.absolute(canny_img), "gray")
plt.show()

# np absolute biar hasilnya cuma b&w, ndak ada value nanggung
```
