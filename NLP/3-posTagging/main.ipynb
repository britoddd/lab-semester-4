# **POS Tagging**

*Makrolinguistik* -> dalam skala luas, psikologi, sosiologi, dll\
*Mikrolinguistik* -> fokus ke struktur dari bahasa itu sendiri: fonologi, morfologi, **sintaksis (ini POS tagging)**, dkk

POS = Part of Speech (jenis/kelas kata: noun, pronoun, adj, etc)\
Tagging = Proses memberi label tiap kata

```python
# Cleansing
import string  # alternatif dari regex
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer  # pas kuis bakal antara lemmatizing/stemming
from nltk.tag import pos_tag  # dipake di Google Translate, biar tau struktur dari katanya, ngga raw translate
from nltk.tokenize import word_tokenize
from nltk import FreqDist  # ngitung frekuensi untuk setiap kata yang ada
```

```python
punctuation = set(string.punctuation)
eng_stop_words = set(stopwords.words('english'))  # kalo jadi set bakal lebih bagus, karena hashing, default aslinya masih list
lemmatizer = WordNetLemmatizer()
```

```python
def get_tag(tag):  # dibuat untuk menyesuaikan function lemmatize yang ada di bawah
    if tag.startswith('JJ'):
        return 'a'
    elif tag.startswith('VB'):
        return 'v'
    elif tag.startswith('RB'):
        return 'r'
    else:
        return 'n'

def lemmatize(tokens):
    lemmatized = []
    pos = pos_tag(tokens)
    for word, tag in pos:
        lemmatized.append(lemmatizer.lemmatize(word, pos=get_tag(tag)))
    return tokens
```

```python
def preprocess(sentence):
    sentence = sentence.lower()
    tokens = word_tokenize(sentence)
    tokens = [token for token in tokens if token not in punctuation]
    tokens = [token for token in tokens if token not in eng_stop_words]
    tokens = lemmatize(tokens)
    return tokens
```

```python
sentence = 'I will read a book.'
print(preprocess(sentence))

freq = FreqDist(sentence)
print(freq.most_common(3))
```

```python
pos_tag(word_tokenize(sentence))  # proses dari pos tagging bisa cek di SS 3 & 4, udah ada strukturnya
```
