**Stemming**\
cuma motong suku katanya, jadi bisa jadi ada karakter yang hilang
dia bakal bandingin beberapa kata, lalu cari irisan yang sama sebagai hasil, jadi akurasinya jelek

Ex:
- continuing -> continu(ing)
- caresses -> caress(es)

Types of stemmers
1. Porter Stemmer -> hanya bisa di bahasa inggris
2. Snowball Stemmer -> bisa di banyak bahasa, porter 2
3. Lancaster Stemmer -> paling aggresive, paling cepat, tapi hasilnya jelke banget

```python
from nltk.stem import PorterStemmer, SnowballStemmer, LancasterStemmer
import pandas as pd
```

```python
porter = PorterStemmer()
snowball = SnowballStemmer('english')
lancaster = LancasterStemmer()
```

```python
words = [
    "studies", "studying", "better", "children", "mice", "running", "happiness", "relational", "organization", "went", "universe", "university"
]
```

```python
data = []
for word in words:
    data.append({
        "Original": word,
        "Porter": porter.stem(word),
        "Snowball": snowball.stem(word),
        "Lancaster": lancaster.stem(word),
    })
    
df = pd.DataFrame(data)
df
```


**Lemmatizing**\
mencari kata dasarnya juga, jadi bakal kepisah dengan lebih baik
bakal akses Lexical database terlebih dahulu menemukan kata asli

Cons: bakal makan waktu lebih lama

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
for word in words:
    print(word, "->", lemmatizer.lemmatize(word))
    

# untuk noun, preposition-nya nggak akan diilangin (makanya masih ada '-ing'nya)
```

```python
word = "better"

adverb = lemmatizer.lemmatize(word, "r")
verb = lemmatizer.lemmatize(word, "v")
adjective = lemmatizer.lemmatize(word, "a")
noun = lemmatizer.lemmatize(word)  # by default bakal udah noun, ngga usah diset lagi

print(adverb, verb, adjective, noun)
```

```python
from nltk.tag import pos_tag
from nltk.corpus import wordnet
```

```python
pos_tags = pos_tag(words)
pos_tags
```

```python
def get_pos(tag):
    if tag.startswith("J"):
        return wordnet.ADJ
    elif tag.startswith("R"):
        return wordnet.ADV
    elif tag.startswith("V"):
        return wordnet.VERB
    return wordnet.NOUN
```

```python
for word, tag in pos_tags:
    print(word, '->', lemmatizer.lemmatize(word, get_pos(tag)))
```

```python
data = []
for word, tag in pos_tags:
    data.append({
        "Original": word,
        "Porter": porter.stem(word),
        "Snowball": snowball.stem(word),
        "Lancaster": lancaster.stem(word),
        "Lemmatizer": lemmatizer.lemmatize(word),
        "Lemmatizer (POS)": lemmatizer.lemmatize(word, get_pos(tag))
    })
    
df = pd.DataFrame(data)
df
```

## WordNet
- lexical database untuk bahasa inggris

disimpan dalam bentuk synsets (synonim sets), isinya:
- synonyms
- anonyms
- definition
- hypernym & hyponym\
*ex: cat is an animal (cat is hypernym)*

```python
from nltk.corpus import wordnet
```

```python
for word in words:
    synonyms = []
    antonyms = []
    
    synsets = wordnet.synsets(word)
    
    for syn in synsets:
        for lemma in syn.lemmas():
            synonyms.append(lemma.name())
            for ant in lemma.antonyms():
                antonyms.append(ant.name())
                
    print(f"Words: {words[:5]}")
    print(f"Synonyms: {synonyms[:5]}")
    print(f"Antonyms: {antonyms[:5]}")
    print("--------------------")
```
