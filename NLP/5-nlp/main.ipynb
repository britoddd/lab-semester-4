```python
import os
import pandas as pd
from nltk.tokenize import word_tokenize  # yah, intinya ini sambi nginget urutan kerjaannya aja
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tag import pos_tag
from nltk.probability import FreqDist
from nltk.classify import NaiveBayesClassifier, accuracy
import pickle
import string

MODEL_PATH = './model.pickle'
DATA_PATH = './financial_dataset.csv'
```

```python
stopwords = stopwords.words('english')
lemmatizer = WordNetLemmatizer()

# def train_model():
    
```

```python
def get_tag(tag):
    if tag.startswith('J'):  # adjective
        return 'a'
    if tag.startswith('R'):
        return 'r'
    if tag.startswith('V'):
        return 'v'
    if tag.startswith('J'):
        return 'n'

def lemmatizing(tokens):
    lemmatized = []
    tagged = pos_tag(tokens)
    for word, tag in tagged:
        result = lemmatizer.lemmatize(word, get_tag(tag))
        lemmatized.append(result)
    return lemmatized
    
def preprocess_sentence(sentence):
    sentence = sentence.lower()
    tokens = word_tokenize(sentence)
    tokens = [token for token in tokens if token not in stopwords]
    tokens = [token for token in tokens if token.isalpha()]  # buat ilangin punctuation, semua yang nggak alphanumeric bakal ilang
    tokens = lemmatizing(tokens)
    return tokens
```
