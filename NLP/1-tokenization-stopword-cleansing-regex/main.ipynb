# **Tokenization**

*Token*: unit terkecil dari teks (bisa berupa kata, kalimat, karakter, simbol)\
*Tokenization*: menghasilkan token dari teks

Jenis-jenis tokenization:
1. *Sentence Tokenizaiton* -> untuk bikin *summary* dari suatu buku/paper
2. *Word Tokenization* -> untuk analisis *review* (ex: review Gmaps)\
Model akan bisa mempelajari pola, memberikan label dari kata-kata
3. *Character Tokenization* -> dipake untuk cek typo
4. *SubWord Tokenization* -> ini yang dipake sama LLM, model akan bisa mempelajarin kata-kata baru dengan lebih mudah\
Ex: dari kata dasar 'lari', model bisa paham lebih kurang dari kata 'ber-lari-an' atau 'ber-lari' (kalo pake word ga bakal dipahamin sama model). Tau mecahnya gimana, langsung udah ada di *library* biasanya

### **Tokenization Examples:**
*Kalimat*: "Saya hari ini belajar NLP. Ini lumayan seru."

dengan *Sentence Tokenization* menjadi ['Saya hari ini belajar NLP.", "Ini lumayan seru.']\
dengan *Word Tokenization* menjadi ['Saya', 'hari', 'ini', 'belajar', 'NLP', '.,' ...]\
dengan *Char Tokenization* menjadi ['S', 'a', 'y', 'a', ' ', ...]\
dengan *SubWord Tokenization* menjadi ['Sa', 'ya', ' ', ...]\

```python
# Sentence Tokenization

from nltk import sent_tokenize

text = 'Saya hari ini belajar NLP. Ini lumayan seru.'
sent_tokenize(text)
```

```python
# Word Tokenization

from nltk.tokenize import word_tokenize

text = 'Saya hari ini belajar NLP. Ini lumayan seru.'
word_tokenize(text)

# kalo pake text.split(), titiknya bakal nyatu sama NLP, karena split itu ngeliat ' '
text.split()
```

# **StopWords Removal**

**Text Data Hierarchy**\
*Dokumen* -> isinya teks buanyak (ibarat: halaman buku)\
*Corpus* -> banyak dokumen (ibarat: buku)\
*Corpora* -> corpus tapi jamak (ibarat: banyak buku)

Untuk data kecil, StopWords disimpan dalam bentuk list, jadi akan iterasi satu per satu **O(n)**
-> tentu tidak efektif untuk data yang jumlahnya besar

Lebih efektif kalo disimpen *as a **set*** yang udah pake *hash table* **O(1)**, pake ini bisa bikin *custom stopwords* sendiri juga

**Apakah wajib dilakukan?**\
Ngga mesti, tergantung kebutuhannya aja. Di zaman sekarang bisa pake token yang ngasih *weight* untuk tiap kata.

**Apa tujuan awal dari *StopWords*?**\
Biar enteng, karena ngeproses StopWords akan ngga berguna + bikin berat-beratin aja

```python
from nltk.corpus import stopwords

ind_stop_words = set(stopwords.words('indonesian'))  # kudu di TypeCast kalo mau pake set, pas Quiz pakenya English
# Notes banget: ini ngga best practice untuk bahasa indonesia, pake Sastrawi aja kalo Indo

custom_stop_words = ind_stop_words - {'saya'}  # buat remove suatu token dari set
print(ind_stop_words)
```

```python
text = 'Saya hari ini belajar NLP. Ini lumayan seru.'

filtered_stopwords = []
words = word_tokenize(text)
print("Before:", words)

for word in words:
    if word.lower() not in ind_stop_words:
        filtered_stopwords.append(word)
print("After:", filtered_stopwords)

custom_filtered_stopwords = []
for word in words:
    if word.lower() not in custom_stop_words:
        custom_filtered_stopwords.append(word)
print("After (custom):", custom_filtered_stopwords)
```

# **Cleansing** (hapus angka & simbol)

ini ngga wajib dilakukan (ngga ada *something* yang wajib dalam dunia NLP)
pakenya *Regular Expression*

```python
import re

text = 'Saya belajar NLP di tahun 2026.'
# re.sub(pola yang ingin dihilangkan, ganti jadi apa dari hasil search polanya, text yang mau diolah pake RegEX)  
# sub = substitute

re.sub(r'\d', '', text)
re.sub(r'[0-9]', '', text)
re.sub(r' ', '', text)

# di sini, 'r' kepake buat 'ngecancel' regular expression, contoh r'\n' bakal mencegah Enter
```

```python
import re

text = 'Saya belajar NLP di tahun 2026. 😊'
text = re.sub(r"[^\w\s]", '', text)  # both backslash kena negation
text = re.sub(r"[^\w^\s]", '', text)  # another practice
text = re.sub(r"[^a-zA-Z0-9_ -]", '', text)  # another practice, pake - bisa substitute fungsi strip

text.strip()  # lebih bagus, buat ngilangin character yang ngga perlu
```
