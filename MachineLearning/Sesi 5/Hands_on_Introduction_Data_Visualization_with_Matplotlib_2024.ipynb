```python
#Load a comma separated file (CSV file) into a DataFrame:
import pandas as pd

df = pd.read_csv("https://media.geeksforgeeks.org/wp-content/uploads/tips.csv")

display(df.head(10))
```

# Scatter plot using matplotlib

```python
import matplotlib.pyplot as plt
 
# Scatter plot with day against tip
plt.scatter(df['day'], df['tip'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
 
plt.show()
```

```python
# Scatter plot with day against tip
plt.scatter(df['day'], df['tip'], c=df['size'], 
            s=df['total_bill'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
 
plt.colorbar()
```

```python
# Line Chart with day against tip
plt.plot(df['tip'])
plt.plot(df['day'])
 
# Adding Title to the Plot
plt.title("Line Chart")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
 
plt.show()
```

```python
# Bar chart with day against tip
plt.bar(df['day'], df['tip'])
 
plt.title("Bar Chart")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')

# Adding the legends
plt.show()
```

```python
# histogram of total_bills
plt.hist(df['total_bill'])
 
plt.title("Histogram")
# Setting the X and Y labels
plt.ylabel('USD Amount')
plt.xlabel('Total Bill') 
# Adding the legends
plt.show()
```

```python
# create histogram for numeric data
df.hist()
 
# show plot
plt.show()
```
