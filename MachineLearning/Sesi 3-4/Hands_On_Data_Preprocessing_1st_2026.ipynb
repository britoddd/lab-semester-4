```python
#load library
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
#Importing the Dataset
dataset = pd.read_csv("https://raw.githubusercontent.com/joh4nnes/MLData/main/purchase.csv", encoding='latin-1')
```

```python
print(dataset.head())
```

### X = independent variables and y = dependent variable

```python
X = dataset.iloc[1, [0,1,2]].values
y = dataset.iloc[:, 3].values 
```

```python
# Print Independent Variables
print("X:")
print(X)
```

```python
# Print Dependent Variables
print("Y:")
print(y)
```

### Handling the missing values

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

```python
X
```

### Encoding the categorical variables

```python
#Encoding the categorical variables
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
labelencoder_X = LabelEncoder()
X[:, 0] = labelencoder_X.fit_transform(X[:, 0])
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
y = y.reshape(-1, 1)
```

```python
print(y)
```

### Creating dummy variables for the encoded categorical values

```python
#Creating dummy variables for the encoded categorical values
transformer = ColumnTransformer([('one_hot_encoder', OneHotEncoder(), [0])],remainder='passthrough')
X = np.array(transformer.fit_transform(X), dtype=np.float64)
```

```python
X
```

### Splitting the datasets

```python
#Splitting the datasets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0 )
```

### Feature Scaling

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train)
```
