## Line Chart

```python
import plotly.express as px
 
# using the iris dataset
df = px.data.iris()
 
# plotting the line chart
fig = px.line(df, y="sepal_width", line_group='species')
 
# showing the plot
fig.show()
```

## Bar Chart

```python
#Load a comma separated file (CSV file) into a DataFrame:
import pandas as pd

df = pd.read_csv("https://media.geeksforgeeks.org/wp-content/uploads/tips.csv")

display(df.head(10))
```

```python
import plotly.express as px

# Creating the bar chart
fig = px.bar(df, x='day', y="total_bill")
 
fig.show()
```

```python
# Creating the bar chart
fig = px.bar(df, x='day', y="total_bill", color='sex',
             facet_row='time', facet_col='sex')
 
fig.show()
```

## Scatter plot using Plotly

```python
import plotly.express as px
 
 
# plotting the scatter chart
fig = px.scatter(df, x='total_bill', y="tip")
 
# showing the plot
fig.show()
```

```python
# plotting the scatter chart
fig = px.scatter(df, x='total_bill', y="tip", color='time',
                 symbol='sex', size='size', facet_row='day',
                 facet_col='time')
 
# showing the plot
fig.show()
```

## Histogram

```python
# plotting the histogram

fig = px.histogram(df, x="total_bill",title="Tips Count")
 
# showing the plot
fig.show()
```

```python
# plotting the histogram
fig = px.histogram(df, x="total_bill", color='sex',
                   nbins=50, histnorm='percent',
                   barmode='overlay')
 
# showing the plot
fig.show()
```

## Pie Chart

```python
fig = px.pie(df, values="total_bill", names="day")
fig.show()
```

```python
fig = px.pie(df, values="total_bill", names="day",
             color_discrete_sequence=px.colors.sequential.RdBu,
             opacity=0.7, hole=0.5)
fig.show()
```
